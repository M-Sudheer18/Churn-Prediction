Accuracy -- Measures overall correctness

Confusion Matrix -- True Positive (TP), False Positive (FP), True Negative (TN), False Negative (FN).

Classification Report -- Precision, Recall, F1-score, Support.

Precision -- Out of predicted positives, how many are correct

Recall -- Out of actual positives, how many you captured

F1 Score -- Balance between precision & recall

```python
# For Accuracy, Confusion Matrix, Classification Report

def train1(metric, metric_name, y_train, y_pred_train):
  result = metric(y_train, y_pred_train)
  print(f"{metric_name}: {result}")


def test1(metric, metric_name, y_test, y_pred_test):
  result = metric(y_test, y_pred_test)
  print(f"{metric_name}: {result}")
```

```python
# For Precision, Recall, F1 Score

def met_tr(metric, metric_name, y_train, y_pred_train, average = "weighted"):
  result = metric(y_train, y_pred_train, average=average)
  print(f"{metric_name}: {result}")

def met_te(metric, metric_name, y_test, y_pred_test, average = "weighted"):
  result = metric(y_test, y_pred_test, average=average)
  print(f"{metric_name}: {result}")
```

```python
from sklearn.metrics import accuracy_score

train1(
    accuracy_score,
    "acc_score",
    y_train,
    y_pred_train
)

train1(
    accuracy_score,
    "acc_score",
    y_test,
    y_pred_test
)
```

```python
from sklearn.metrics import confusion_matrix

train1(
    confusion_matrix,
    "cm_train",
    y_train,
    y_pred_train
)

train1(
    confusion_matrix,
    "cm_test",
    y_test,
    y_pred_test
)
```

```python
from sklearn.metrics import classification_report

train1(
    classification_report,
    "cr_train",
    y_train,
    y_pred_train
)

train1(
    classification_report,
    "cr_test",
    y_test,
    y_pred_test
)
```

```python
from sklearn.metrics import precision_score

met_tr(
    precision_score,
    "ps_train",
    y_train,
    y_pred_train,
    average = 'weighted'
)

met_te(
    precision_score,
    "ps_test",
    y_test,
    y_pred_test,
    average = 'weighted'
)
```

```python
from sklearn.metrics import f1_score

met_tr(
    f1_score,
    "f1_train",
    y_train,
    y_pred_train,
    average = 'weighted'
)

met_te(
    f1_score,
    "f1_test",
    y_test,
    y_pred_test,
    average = 'weighted'
)
```

```python
from sklearn.metrics import recall_score

met_tr(
    recall_score,
    "rs_train",
    y_train,
    y_pred_train,
    average = 'weighted'
)

met_te(
    recall_score,
    "rs_test",
    y_test,
    y_pred_test,
    average = 'weighted'
)
```
